```python
# 2. Make a new DataFrame of listings in Brooklyn named `df_bk` 
# and find how many listings in just Brooklyn.

brooklyn_listings_condition = df.neighbourhood_group == 'Brooklyn'
df_bk = df[brooklyn_listings_condition]
print('Listings in just Brooklyn:', len(df_bk))
df_bk
```

```python
df.head()
```

```python
# 4. Using `.isin()` select anyone that has the host name of Michael, David, John, and Daniel.
# How many total are there that have those names

host_names = ["Michael","David","John","Daniel"]

#making boolean mask using isin() for selecting statement
condition = df.host_name.isin(host_names)

#displaying the filtered data
print(f"There are {len(df[condition])} total are there that have those names of Michael, David, John, and Daniel")
df[condition]["host_name"].value_counts()
```

```python
df.head()
```

```python
# 5. Create a new column called `adjusted_price` that has $100 added to every listing in Williamsburg.  
# The prices for all other listings should be the same as the were before. 

condition1 = df['neighbourhood'] == 'Williamsburg'
condition2 = df['price'] > 100

df['adjusted_price'] = np.where(condition1 & condition2, df['price'] + 100, df["price"])
df

# Sanity check to view if our assignment worked correctly.
select_statement = df['adjusted_price'] > df['price']
new_df = df[  select_statement ]
df['select_statement'] = select_statement

df.head()
```

```python
# 6. What % of the rooms are private, and what % of the rooms are shared.  

percentage_of_private = df['room_type'].value_counts(normalize= True)['Private room']
percentage_of_shared = df['room_type'].value_counts(normalize= True)['Shared room']

print(f"{percentage_of_private}% of the rooms are private")
print(f"{percentage_of_shared}% of the rooms are shared")


```

# Exercise 3 - Grouping

1. Using `groupby`, count how many listings are in each neighbourhood_group.


2. Using `groupby`, find the mean price for each of the neighbourhood_groups. 


3. Using `groupby` and `.agg()`, find the min and max price for each of the neighbourhood_groups. 


4. Using `groupby`, find the median price for each room type in each neighbourhood_group.


5. Using `groupby` and `.agg()`, find the count, min, max, mean, median, and std of the prices for each room type in each neighbourhood_group.

```python
# 1. Using `groupby`, count how many listings are in each neighbourhood_group.

grouped = df.groupby('neighbourhood_group')
listing_counts = grouped.size()
print("listings in each neighbourhood_group: ", listing_counts)
```

```python
# 2. Using `groupby`, find the mean price for each of the neighbourhood_groups. 
grouped['price'].mean()

```

```python
# 2.5. Using `groupby`, find the mean price for each room_type. 
room_type_mean_prices = df.groupby('room_type')
print("listings in each neighbourhood_group: ", room_type_mean_prices['price'].mean())
```

```python
# 3. Using `groupby` and `.agg()`, find the min and max price for each of the neighbourhood_groups. 
grouped = df.groupby("neighbourhood_group")
grouped["price"].agg(["min", "max"])
```

```python
# 4. Using `groupby`, find the mean price for each room_type in each neighbourhood_group.
groupby_cols = ['neighbourhood_group', 'room_type']

grouped = df.groupby(groupby_cols)
grouped['price'].mean()
```

```python
# 5. Using `groupby` and `.agg()`, find the count, min, max, mean, median, and std of the prices 
# for each room type in each neighbourhood_group.

grouped['price'].agg(['count', 'min', 'max', 'mean', 'median','std'])
```
